```python
n = int(input())
res = [[0 for i in range(n)]for i in range(n)]
a=1
t,b,r,l = 0,n-1,n-1,0
while (a<=n*n):
    for i in range(l,r+1):
        res[t][i]=a
        a+=1
    t+=1

    for i in range(t,b+1):
        res[i][r]=a
        a+=1
    r-=1

    for i in range(r, l-1, -1):
        res[b][i]=a
        a+=1
    b-=1

    for i in range(b, t-1, -1):
        res[i][l]=a
        a+=1
    l+=1

print(res)
```
